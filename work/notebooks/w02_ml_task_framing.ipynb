## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

## 1. My Lane as an ML Task

I chose the **Content Refresh** lane.

This problem is best framed as a **classification** task. The goal is to classify whether a content page is likely to require a content refresh based on its historical search performance and engagement metrics.

The output helps content teams prioritize which pages should be reviewed first.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
## 2. Target or Proxy

The target is whether a page is showing signs of declining search performance.

Since the starter dataset does not contain a direct refresh label, I use the **trend_direction** column as a proxy. Pages labeled **down** represent declining content, while the remaining values represent stable or improving content.

This proxy is based on observed historical performance and is intended for decision support rather than proving causation.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
## 3. Success Metric

The primary success metric is **Precision**.

High precision means that most pages recommended for review are actually declining, helping content teams spend their time on the highest-priority pages.

Recall is also important because missing declining pages could result in continued traffic loss.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
## 4. The Unit of Analysis

Each row in the dataset represents **one content page**.

The row contains search performance, engagement metrics, content characteristics, and trend information that can be used to support refresh decisions.
```
